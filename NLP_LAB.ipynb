EXP 01 - Write program demonstrates how to use regular expressions in Python to match and search for patterns in text


```python
import re

text = "My email is student123@gmail.com and my phone number is 9876543210."

email = re.search(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b', text)

if email:
    print("Email found:", email.group())
else:
    print("Email not found.")

match = re.match(r"My", text)

if match:
    print("Matched at the beginning:", match.group())
else:
    print("No match at the beginning.")

phone = re.search(r'\b\d{10}\b', text)

if phone:
    print("Phone number found:", phone.group())
else:
    print("Phone number not found.")
```

EXP 02- Implement a basic finite state automaton that recognizes a specific language or pattern. In this
example, we'll create a simple automaton to match strings ending with 'ab' using python


```python
def fsa(string):
    state = 0

    for ch in string:
        if state == 0:
            if ch == 'a':
                state = 1
            else:
                state = 0
        elif state == 1:
            if ch == 'a':
                state = 1
            elif ch == 'b':
                state = 2
            else:
                state = 0
        elif state == 2:
            if ch == 'a':
                state = 1
            else:
                state = 0

    if state == 2:
        print("Accepted")
    else:
        print("Rejected")

s = input("Enter a string: ")
fsa(s)
```

EXP 03 -Write program demonstrates how to perform morphological analysis using the NLTK library in
Python

```python
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer

nltk.download('wordnet')

ps = PorterStemmer()
lm = WordNetLemmatizer()

words = ["running", "studies", "better", "playing", "cats"]

print("Word\t\tStem\t\tLemma")

for word in words:
    print(word, "\t", ps.stem(word), "\t\t", lm.lemmatize(word))
```

EXP 04- Implement a finite-state machine for morphological parsing. In this example, we'll create a simple machine to generate plural forms of English nouns using python.

```python
def plural(noun):
    if noun.endswith(("s", "x", "z", "ch", "sh")):
        return noun + "es"
    elif noun.endswith("y") and noun[-2] not in "aeiou":
        return noun[:-1] + "ies"
    else:
        return noun + "s"

word = input("Enter a noun: ")
print("Plural form:", plural(word))
```

EXP 06-Implement a basic N-gram model for text generation. For example, generate text using a bigram
model using python

```python
import random

text = "the cat sat on the mat the cat ate the fish"

words = text.split()

bigrams = {}

for i in range(len(words) - 1):
    key = words[i]
    value = words[i + 1]
    if key not in bigrams:
        bigrams[key] = []
    bigrams[key].append(value)

word = random.choice(words)
result = [word]

for i in range(10):
    if word in bigrams:
        word = random.choice(bigrams[word])
        result.append(word)
    else:
        break

print("Generated Text:")
print(" ".join(result))
```

EXP 07-Write program using the NLTK library to perform part-of-speech tagging on a text.

```python
import nltk
from nltk import word_tokenize, pos_tag

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng') # Added this line to download the missing resource

text = "The cat is sitting on the mat."

words = word_tokenize(text)
tags = pos_tag(words)

print("Part-of-Speech Tags:")
for word, tag in tags:
    print(word, ":", tag)
```

EXP 08 -Implement a simple stochastic part-of-speech tagging algorithm using a basic probabilistic
model to assign POS tags using python.

```python
import random

tag_prob = {
    "the": ["DT"],
    "a": ["DT"],
    "cat": ["NN"],
    "dog": ["NN"],
    "runs": ["VBZ"],
    "eats": ["VBZ"],
    "fast": ["RB"],
    "quickly": ["RB"]
}

sentence = input("Enter a sentence: ").lower().split()

print("POS Tags:")

for word in sentence:
    if word in tag_prob:
        tag = random.choice(tag_prob[word])
    else:
        tag = "NN"
    print(word, ":", tag)
```

EXP 09-Implement a rule-based part-of-speech tagging system using regular expressions using python

```python
import nltk
from nltk import word_tokenize
from nltk.tag import RegexpTagger

nltk.download('punkt')

patterns = [
    (r'.*ing$', 'VBG'),
    (r'.*ed$', 'VBD'),
    (r'.*ly$', 'RB'),
    (r'.*ous$', 'JJ'),
    (r'.*s$', 'NNS'),
    (r'^[0-9]+$', 'CD'),
    (r'.*', 'NN')
]

tagger = RegexpTagger(patterns)

text = "The boys are playing happily in the garden"

words = word_tokenize(text)
tags = tagger.tag(words)

print("POS Tags:")
for word, tag in tags:
    print(word, ":", tag)
```
